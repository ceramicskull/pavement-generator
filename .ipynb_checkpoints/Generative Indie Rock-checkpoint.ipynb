```python
import pandas as pd
import numpy as np
import math

signalL = signalData[:,0]
signalR = signalData[:,1]
sequence_length = 100
len(signalL)
network_input = np.empty((len(signalL)-sequence_length,sequence_length,2),dtype="int16")
network_output = np.empty((len(signalL)-sequence_length,2),dtype="int16")
for i in range(len(signalL) - sequence_length):
    sequence_in_left = signalL[i:i+sequence_length]
    sequence_in_right = signalR[i:i+sequence_length]
    sequence_out_left = signalL[i+sequence_length]
    sequence_out_right = signalR[i+sequence_length]
    network_input[i,:,0] = sequence_in_left
    network_input[i,:,1] = sequence_in_right
    network_output[i,0] = sequence_out_left
    network_output[i,1] = sequence_out_right
    
print(network_input[0])
    
```

```python
print(network_input[4000000])
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.callbacks import ModelCheckpoint
model = Sequential()
model.add(LSTM(
        256,
        input_shape=(network_input.shape[1], network_input.shape[2]),
        return_sequences=True
    ))
model.add(Dropout(0.3))
model.add(LSTM(512, return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(256))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(2))
model.add(Activation('softmax'))
model.compile(loss='mean_squared_error', optimizer='rmsprop')

filepath = "weights-improvement-{epoch:02d}-{loss:.4f}-bigger.hdf5"    
checkpoint = ModelCheckpoint(
    filepath, monitor='loss', 
    verbose=0,        
    save_best_only=True,        
    mode='min'
)    
callbacks_list = [checkpoint]     
model.fit(network_input, network_output, epochs=200, batch_size=64, callbacks=callbacks_list)
```
